```python
# multivariate regression
# predicting home prices based on area bedrooms age of home
# price=m1*area+m2*bedrooms+m3*age+b
```

```python
df = pd.read_csv("/kaggle/input/home-price-prediction-multi-feature/homeprices (2).csv")
df
```

```python
median_no= df.bedrooms.median().astype('int64')
median_no
```

```python
df.bedrooms = df.bedrooms.fillna(median_no)
df
```

```python
#creating a linear regression class object
#then use fit method with independent variables: area bedrooms age
#dependent variable price
reg=linear_model.LinearRegression()
reg.fit(df[['area','bedrooms','age']],df.price)
```

```python
# M1 M2 M3
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000,3,40]])
```
