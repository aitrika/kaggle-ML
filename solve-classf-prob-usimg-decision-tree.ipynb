```python
df = pd.read_csv("/kaggle/input/salaries-dataaset/salaries.csv")
df
```

# separate dependent and independent variables

```python
input = df.drop('salary_more_then_100k',axis='columns')
target = df.salary_more_then_100k
input
```

# changing categorical data to numerical

```python
from sklearn.preprocessing import LabelEncoder
le_comp = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
input['comp_n'] = le_comp.fit_transform(input['company'])
input['job_n'] = le_job.fit_transform(input['job'])
input['degree_n'] = le_degree.fit_transform(input['degree'])
input
```

```python
input_n = input.drop(['company','job','degree'],axis=1)
input_n
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(input_n,target)
model.predict([[2,0,1]])
```

# predict survival rate of passengers in titanic

```python
hw = pd.read_csv("/kaggle/input/titanic/titanic.csv")
hw
```

```python
target = hw['Survived']
target
hw_final = hw.drop(['PassengerId','Survived','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
hw_final
```

# replace NaN values in Age column

```python
median_val = hw_final['Age'].median()
hw_final['Age'] = hw_final['Age'].fillna(median_val)
hw_final
```

```python
from sklearn.preprocessing import LabelEncoder
le_sex = LabelEncoder()
hw_final['sex_n']= le_sex.fit_transform(hw_final['Sex'])
hw_final = hw_final.drop('Sex',axis='columns')
hw_final
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(hw_final,target)
model.predict([[3,26,7.2500,0]])
```
